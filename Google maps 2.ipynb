```python
try:
    time.sleep(5)

    scrollable_div = driver.find_element(By.CSS_SELECTOR, 'div[role="feed"]')
    driver.execute_script("""
        var scrollable_div = arguments[0];
        function scrollWithinElement(scrollableDiv) {
            return new Promise((resolve, reject) => {
                var totalHeight = 0;
                var distance = 1000;
                var scrollDelay = 3000;

                var timer = setInterval(() => {
                    var scrollHeightBefore = scrollable_div.scrollHeight;
                    scrollable_div.scrollBy(0, distance);
                    totalHeight += distance;

                    if (totalHeight >= scrollHeightBefore) {
                        totalHeight = 0;
                        setTimeout(() => {
                            var scrollHeightAfter = scrollable_div.scrollHeight;
                            if (scrollHeightAfter <= scrollHeightBefore) {
                                clearInterval(timer);
                                resolve();
                            }
                        }, scrollDelay);
                    }
                }, 200);
            });
        }
        return scrollWithinElement(scrollable_div);
    """, scrollable_div)

    time.sleep(30)
    
    items = driver.find_elements(By.CSS_SELECTOR, 'div[role="feed"] > div > div[jsaction]')

    results = []

    for item in items:
        data = {}
        
        try:
            data['title'] = item.find_element(By.CSS_SELECTOR, '.fontHeadlineSmall').text
            if data.get('title'):
                results.append(data)
        except:
            pass

        try:
            data['link'] = item.find_element(By.CSS_SELECTOR, 'div[role="feed"] > div > div[jsaction] div > a').get_attribute('href')
            if data.get('link'):
                results.append(data)
        except:
            pass

    df = pd.DataFrame(results)
    print(df)

finally:
    time.sleep(29)
    driver.quit()
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.parse
from collections import deque
import re
import numpy as np

# Assuming df is already defined and contains a 'link' column
links = df['link'].dropna().astype(str).tolist()  # Drop NaN values and convert to string
urls = deque(links)
scraped_urls = set()

# Dictionary to store emails for each link
link_emails = {link: [] for link in links}

count = 0
try:
    while len(urls):
        count += 1
        if count > 100:
            break
        url = urls.popleft()
        if url in scraped_urls:
            continue
        scraped_urls.add(url)
        
        # Ensure url is a string
        url = str(url)
        
        parts = urllib.parse.urlsplit(url)
        base_url = '{0.scheme}://{0.netloc}'.format(parts)
        path = url[:url.rfind('/') + 1] if '/' in parts.path else url
        print('[{}] Processing {}'.format(count, url))
        try:
            response = requests.get(url)
        except (requests.exceptions.MissingSchema, requests.exceptions.ConnectionError, requests.exceptions.InvalidURL, requests.exceptions.RequestException):
            continue

        new_emails = set(re.findall(r'[a-z0-9.\-_+]+@[a-z0-9.\-_]+\.[a-z]+', response.text, re.I))
        
        # Use setdefault to initialize the list if url is not in link_emails
        link_emails.setdefault(url, []).extend(new_emails)

        soup = BeautifulSoup(response.text, 'html.parser')

        for anchor in soup.find_all("a"):
            link = anchor.get('href', '')
            if link.startswith('/'):
                link = urllib.parse.urljoin(base_url, link)
            elif not link.startswith('http'):
                link = urllib.parse.urljoin(path, link)
            if link not in urls and link not in scraped_urls:
                urls.append(link)
except KeyboardInterrupt:
    print('[-] Closing')

# Create a list to store combined emails for each link
emails_combined = []

# Iterate through links in the DataFrame and add combined emails
for link in df['link']:
    link = str(link)  # Ensure link is a string
    emails_combined.append(', '.join(link_emails.get(link, [])))  # Use get() to handle missing keys

# Add emails to DataFrame
df['Emails'] = emails_combined

# Print or further process the emails
for emails in emails_combined:
    print(emails)
```
