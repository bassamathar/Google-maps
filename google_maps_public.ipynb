```python
results = []
wait = WebDriverWait(driver, 20)  # Increased wait time to 20 seconds
counter = 0
max_pages = 10  # Maximum number of pages to scrape

while counter < max_pages:
    items = driver.find_elements(By.CLASS_NAME, 'VkpGBb')

    for item in items:
        data = {}
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        try:
            link_element = item.find_element(By.XPATH, './/a[@class="yYlJEf Q7PwXb L48Cpd brKmxb"]')
            data['link'] = link_element.get_attribute('href')
        except:
            pass

        # Attempt to extract title text
        try:
            title_element = item.find_element(By.CSS_SELECTOR, '.dbg0pd')
            data['title'] = title_element.text
        except Exception as e:
            data['title'] = None

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        results.append(data)

    try:
        print("Trying to find next page element by XPATH...")
        next_page = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[3]/div/div[8]/div[1]/div/div[2]/div[2]/div/div/div/div/div/div/div/div/div[2]/div/table/tbody/tr/td[12]/a')))
        next_page.click()
        counter += 1
    except (TimeoutException, NoSuchElementException, StaleElementReferenceException):
        try:
            print("Trying to find next page element by ID...")
            next_page = wait.until(EC.element_to_be_clickable((By.ID, 'pnnext')))
            next_page.click()
            counter += 1
        except (TimeoutException, NoSuchElementException, StaleElementReferenceException):
            try:
                print("Trying to find next page element by CLASS_NAME...")
                nextt = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'd6cvqb.BBwThe')))
                nextt_page = nextt.find_element(By.TAG_NAME, 'a')
                nexttss = nextt_page.get_attribute('href')
                driver.get(nexttss)
                counter += 1
            except (TimeoutException, NoSuchElementException, StaleElementReferenceException):
                print("No more pages found or error occurred.")
                break

 
# Create DataFrame from results
df = pd.DataFrame(results)
print(df)
```
